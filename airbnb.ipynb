```python
##first,  tokenize the sentences into words a (Apr 2019)
import re
def count_covid(data):
    text = data.dropna()
    text['words'] = text['name'].str.strip().str.split('[\W_]+')

    # second,create new DataFrame with words
    rows = list()
    for row in text[['words']].iterrows():
        r = row[1]
        for word in r.words:
            rows.append((word))

    words = pd.DataFrame(rows, columns=['word'])
    words.head()

    #third, remove empty strings
    words = words[words.word.str.len() > 0]

    #fourht, change to lower cases 
    words['word'] = words.word.str.lower()
    
    selected_words=['covid19','covid', 'Covid-19', 'covid-19', 'COVID19','quarantine', 'corona','coronavirus', 'virus', 'pandemic',\
                    'Health', 'Healthcare', 'Health care', 'hospital', 'hospitals', 'healthcare','health', 'healthy', 'health care',  'frontline']
    res = words['word'].value_counts().reindex(selected_words)
    return res
```
